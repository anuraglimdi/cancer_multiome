# Exploring multi-omic cancer datasets

Specifically looking at number of cell line models, number of features for different datasets, and getting a sense of variation in the different measurements

```python
import os
import pandas as pd
```

```python
cancer_type = "Lung"
```

```python
dataset_dict = {}
for file in os.listdir(f"Dataset/{cancer_type}"):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(f"Dataset/{cancer_type}", file), index_col=0)
        dict_key = file.replace("_imputed", "").replace("_latest", "").replace(".csv", "").split("_")[-1]
        dataset_dict[dict_key] = df
```

variables for accessing each dataframe
